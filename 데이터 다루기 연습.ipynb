### 불량 데이터 처리
#### selloutData 변수에서 QTY컬럼 음수(반품)를 0으로 변경한 값을 QTY_NEW 컬럼에 담으세요
#### kopo_channel_seasonality_new.csv

```python
#먼저 필요한 데이터를 데이터프레임에 담는다.
selloutData = pd.read_csv("./dataset/kopo_channel_seasonality_new.csv")
selloutData.head()
```

```python
#새로운 칼럼을 생성하고 그 값에 조건식을 걸어 나온 결과값을 담는다.
#조건은 0보다 작은 값은 0으로 하고 0보다 작지 않을 경우엔 그대로 두는 것이다.
selloutData["QTY_NEW"] = np.where(selloutData["QTY"]<0,0,selloutData["QTY"])
selloutData.head()
```

##### selloutData 자료에서  YEAR, WEEK 컬럼을 생성하고 WEEK 가 52 이하인 데이터만 조회한 후 WEEK 칼럼에 담으세요

```python
selloutData["YEAR"] = selloutData["YEARWEEK"]/100
#YEAR 칼럼의 내용이 실수가 되었다. 뒤의 실수값을 없애야 한다.
selloutData["WEEK"] = selloutData["YEARWEEK"]%100

# selloutData[["YEAR"]].astype(int)
# selloutData["YEAR"].dtypes
#이렇게 했을 때는 먹히지 않았다. 왜냐하면 인트로 변환한 값을 저장하지 않았기 때문이다.

selloutData["YEAR"] = selloutData[["YEAR"]].astype(int)
selloutData["YEAR"].dtypes
#이런 식으로 다시 값을 담아주어야 한다.
```

```python
selloutData.head()
```

```python
selloutData["WEEK"] = np.where(selloutData["WEEK"] > 52, 0, selloutData["WEEK"])
```

```python
selloutData["WEEK"].max()
```
