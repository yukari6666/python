```python
import pandas as pd
```

```python
selloutData = pd.read_csv('./dataset/kopo_channel_seasonality_new.csv')
selloutData.head()
```

```python
pivotResult = selloutData.pivot_table(index = ["REGIONID", "PRODUCT"],\
                                     columns = "YEARWEEK", values ="QTY",\
                                     aggfunc="mean")
pivotResult.head()
```

```python
pivotResult2 = pivotResult.reset_index()
pivotResult2.head()
```

```python
unpivotResult = pd.melt(pivotResult2, id_vars=["REGIONID", "PRODUCT"],\
                        var_name="YEARWEEK", value_name="QTY")
unpivotResult.head()
```
