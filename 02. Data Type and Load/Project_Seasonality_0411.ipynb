## seasonData 불러오기

```python
seasonData = pd.read_csv('../dataset/kopo_channel_seasonality_new.csv')
```

### seasonData QTY에서 음수 값 제거하고 NEW_QTY에 담기

```python
seasonData["QTY_NEW"] = np.where(seasonData["QTY"] < 0, 0, seasonData["QTY"])
```

```python
refindData.max()
#WEEK 칼럼이 만들어졌고 그 최대 값이 52임을 확인할 수 있다.
```

### seasonData["YEARWEEK"]를 YEAR와 WEEK로 나눠서 칼럼 생성 담기

```python
seasonData["YEAR"] = seasonData["YEARWEEK"]/100
```

```python
seasonData["WEEK"] = seasonData["YEARWEEK"]%100
```

### year와 week를 자동 추출하는 함수 만들기

```python
def customYear(inputColumn):
    yearValue = inputColumn/100
    return int(yearValue)
#교수님이 주신 예제로 함수를 생성. 연도 구분.
```

```python
def customWeek(inputColumn):
    weekValue = inputColumn%100
    return weekValue
#교수님의 예제로 함수 생성. 주차 구분
```

```python
def myCustomYear(inputColumn):
    yearValue = str(inputColumn)[0:4]
    return int(yearValue)
#내가 만든 함수. 스트링 값으로 변환해 인덱스로 구분.
```

```python
def myCustomWeek(inputColumn):
    weekValue = str(inputColumn)[4: ]
    return int(weekValue)
#내가 만든 함수. 스트링 값으로 변환해 인덱스로 구분.
```

### 교수님 함수를 사용해 보았다

```python
seasonData["YEAR"] = seasonData["YEARWEEK"].apply(customYear)
```

```python
seasonData.head()
```

```python
seasonData["WEEK"] = seasonData["YEARWEEK"].apply(customWeek)
```

```python
seasonData.head()
```

```python
seasonData.dtypes
```

### 내가 만든 함수를 사용해보았다

```python
seasonData["YEAR"] = seasonData["YEARWEEK"].apply(myCustomYear)
```

```python
seasonData["WEEK"] = seasonData["YEARWEEK"].apply(myCustomWeek)
```

## 정렬을 해보았다

## 정렬한 데이타프레임을 인덱스 정리해 보았다.

```python
stepData = sortedData.reset_index(drop=True)
```

# 롤링 함수를 사용해보자.

## 1.기본이동평균 함수
## *mean() = 평균함수 구하기

```python
stepData["MA"] = stepData.QTY_NEW.rolling(window = 15, center = True,\
                                         min_periods = 1).mean()
```

## groupby 사용하기

## merge(join)을 해보았다

```python
mergeData = pd.merge(stepData, groupResult, \
                    on = ["REGIONID", "PRODUCT", "YEAR"], how="left")

# 이전에 에러가 난 이유는 합치는 두 데이타프레임에 같은 이름의 다른 값을 가진 칼럼이 존재해서 그 이름으로 합치려 할 경우 컴퓨터가 제대로
# 구분을 하지 못 해 에러를 일으킨다.
#'[QTY_NEW]' NOT IN INDEX
```

# 그룹바이 함수 만들기

```python
def groupRolling(oneGroup):
    
    indexGroupData = oneGroup.reset_index(drop=True)
    indexGroupData["MA2"] = indexGroupData["MA"].rolling(window = 4,\
                                                              center = True,\
                                                        min_periods = 1).mean()
    return indexGroupData
```

```python
finalResult = mergeData.groupby(groupkey).apply(groupRolling)
```
