## year와 week 값을 자동으로 추출해주는 함수 만들기

```python
# 교수님이 주신 예제를 함수로 만들어보았다.
def customFunction(inputColumn):
    yearValue = inputColumn/100
    return int(yearValue)
```

```python
# 교수님이 주신 예제를 응용해서 week값을 추출하도록 만들어보았다.
def customFunction2(inputColumn):
    weekValue = inputColumn%100
    return weekValue
```

```python
# 나만의 방법으로 year값을 추출하도록 만들어보았다.
def myCustomFunction(inputCoulmn):
    inputCoulmn = ""
    yearValue = inputCoulmn[0:4]
    return int(yearValue)
```

```python
#나만의 방법으로 week값을 추출하도록 만들어보았다.
def myCustomFunction2(inputCoulmn):
    inputCoulmn = ""
    weekValue = inputCoulmn[4: ]
    return weekValue
```

### 교수님의 예제로 만든 함수를 사용해보았다

```python
seasonData["YEAR"] = seasonData["YEARWEEK"].apply(customFunction)
```

```python
seasonData["WEEK"] = seasonData["YEARWEEK"].apply(customFunction2)
```

```python
seasonData.head()
```

### 내가 만든 함수를 사용해보았다

```python
seasonData["YEAR"] = seasonData["YEARWEEK"].apply(myCustomFunction)
```

```python
seasonData["WEEK"] = seasonData["YEARWEEK"].apply(myCustomFunction2)
```
